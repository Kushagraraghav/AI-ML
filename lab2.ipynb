```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("supermarket_sale2.csv")
```

```python
df = pd.read_csv("supermarket_sale2.csv")
df.info()
```

```python
#View the first few rows
df.head()
```

```python
# Get the shape of the dataframe(rows,column)
df.shape
```

```python
# get summary statistics(mean,std,min,max,etc)
df.describe()
```

```python
#check for missing values
df.isnull()
```

```python
#Total missing values
df.isnull().sum()
```

```python
#Drop rows with missing vlues(if appropriate)
df.dropna(inplace=True)
```

```python
df.head()
```

```python
#fill missing values with mean,median or other methods
df.fillna(df['Unit price'].mean(),inplace=True)
```

```python
df.head()
```

```python
# histogram : Visulalize the distribution of numerical variables
sns.histplot(df['Unit price'])
plt.show()
```

```python
#Box plots : Identify outliers and understand the spread of data.
sns.boxplot(x='Payment',y='Unit price',data=df)
plt.show()
```

```python
#Scatter Plots: Examine relationship between two numerical variables.
sns.scatterplot(x='cogs',y='Unit price',data = df)
plt.show()
```

```python
#Bar charts : VIsulize the distribution of categorical variables.
sns.countplot(x='Branch',data=df)
plt.show()
```

```python
# Correlation Matrix : Understand the realtiosnhip betwn multiple numberical variables
af=df[['Unit price','Quantity','Tax 5%','cogs','gross income','Rating']]
corr= af.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
# Group : Analyze based on different categories
df.groupby('City')['Quantity'].mean()
```
