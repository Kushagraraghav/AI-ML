```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("supermarket_sales.csv")
```

```python
data = pd.read_csv("supermarket_sales.csv")
data.info()
```

```python
plt.figure()
plt.plot(data['Unit price'],data['Total'],marker='*')
plt.title('Line Plot')
plt.xlabel('Unit price')
plt.ylabel('Total')
plt.show()
```

```python
plt.figure()
plt.bar(data['Quantity'],data['cogs'])
plt.title('Bar Chart')
plt.xlabel('Quantity')
plt.ylabel('cogs')
plt.show()
```

```python
plt.figure()
plt.scatter(data['Quantity'],data['Unit price'])
plt.title('Scatter Plot')
plt.xlabel('Quantity')
plt.ylabel('Unit price')
plt.show()
```

```python
plt.figure()
sns.histplot(data['Quantity'],bins=6)
plt.title('Histogram')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Quantity',y='Tax 5%',data=data)
plt.title('Box Plot')
plt.xlabel('Quantity')
plt.ylabel('Tax 5 %')
plt.show()
```

```python
af=data[['Unit price','Quantity','Tax 5%','cogs','gross income','Rating']]
corr= af.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, mask=mask, cmap='viridis')
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap='YlGnBu', annot_kws={'size': 12, 'weight': 'bold'}, cbar_kws={'orientation': 'horizontal'})
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap='RdBu_r', center=0)
plt.show()
```

```python
sns.heatmap(corr, annot=True, linewidths=1.5, linecolor='white', cmap='Spectral')
plt.show()
```

```python
mask = np.zeros_like(corr)
mask[corr == 0] = True
sns.heatmap(corr, annot=True, cmap='YlOrBr', mask=mask)
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap='rocket', annot_kws={"bbox": {"boxstyle": "round,pad=0.3", "fc": "yellow", "ec": "black", "lw": 1}})
plt.show()
```
