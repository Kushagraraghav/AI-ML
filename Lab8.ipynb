Implement K- means clustering algorithm using a suitable example in google colab

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data=np.array([
    [1,2],[1.5,1.8],[5,8],[8,8],[1,0.6],[9,11],[8,2],[10,2],[9,3],[3,4],[3.5,5],[4.5,5],[3.8,4.8]
    ])
```

```python
plt.scatter(data[:,0],data[:,1],s=50)
plt.title("Self-Generated Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,random_state=42)
kmeans.fit(data)
```

```python
centers=kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
plt.scatter(data[:,0],data[:,1],c=labels,cmap='viridis',s=50)
plt.scatter(centers[:,0],centers[:,1],c='pink',s=200,marker='X')
plt.title("Kmeans Clustering Resulting (Self_Generated Data)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
